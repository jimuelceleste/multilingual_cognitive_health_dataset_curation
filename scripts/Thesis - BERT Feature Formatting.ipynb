Objective: To format BERT features: mean of tokens.

```python
import os 
import shutil 

import pandas as pd
```

```python
input_dir = '../results/Thesis - BERT Features_last_hidden_state/'
output_dir = '../results/Thesis - BERT Features/'
metadata = os.path.join(input_dir, 'metadata.csv')
metadata_df = pd.read_csv(metadata)
metadata_df.head()
```

```python
%%time
files = metadata_df['filename'].values
n = len(files)
i = 1
for file in files:
    print(f"{i}/{n} {file}")
    input_file = os.path.join(input_dir, file)
    output_file = os.path.join(output_dir, file)
    print(input_file)
    print(output_file)

    if os.path.exists(output_file):
        print("Skipping file: already processed.")
        i += 1
        continue

    # get sum
    embeddings = pd.read_csv(input_file)
    embeddings_mean = embeddings.mean().to_frame().T
    embeddings_mean.to_csv(output_file, index=False)
    
    print("File processed!")
    i += 1
```

```python
metadata_copy = os.path.join(output_dir, 'metadata.csv')
shutil.copy(metadata, metadata_copy)
```
